### 一、张量操作

```python
import torch
import os 
import pandas as pd
```

```python
#张量创建、形状和元素数
x = torch.arange(12)
print(x)
print(x.shape)
print(x.numel())
print(x.reshape(3, 4))
```

```python
# 创建二维列表
x = torch.tensor([[1, 2, 3], [4, 5, 6]])
print(x)
```

```python
# 基本运算
x = torch.tensor([1, 2, 3])
y = torch.tensor([4, 5, 6])
print(x + y)
print(x * y)
print(x / y)
print(x - y)
print(x ** y)
```

```python
x = torch.exp(x)
print(x)
```

```python
# 张量拼接、比较和求和
x = torch.arange(12, dtype=torch.float32).reshape(3, 4)
y = torch.tensor([[2.0, 1.0, 1.0, 0.0], [2.0, 1.0, 1.0, 0.0], [2.0, 1.0, 1.0, 0.0]])
print(x)
print(y)
print(torch.cat((x, y), dim=0))
print(torch.cat((x, y), dim=1))
print(x == y)
print(torch.sum(x))
```

```python
#张量的索引和切片
x = torch.arange(3, dtype=torch.float32)
y = torch.arange(3, dtype=torch.float32)
print(x + y)

print(x[2])
x[2] = 10
print(x)

z = x[1:3]
print(z)
z[:] = 200
print(z)
print(x)

x[:] = x + y
print(x)

```

```python
# 张量与numpy数组转换
a = x.numpy()
print(a)
b = torch.from_numpy(a)
print(b)
```

```python
#标量张量的取值
a = torch.tensor([3.5])
print(a)
print(a.item(), float(a), int(a))
```

#### 二、数据预处理

```python
os.makedirs(os.path.join('..', 'data'), exist_ok=True)
data_file = os.path.join('..', 'data', 'house_tiny.csv')
with open(data_file, 'w') as f:
    f.write('NumRooms,Alley,Price\n')
    f.write('NA,Pave,127500.0\n')
    f.write('2,NA,106000.0\n')
    f.write('4,NA,178100.0\n')
    f.write('NA,NA,140000.0\n')

data = pd.read_csv(data_file)
print(data)
```

```python
# 拆分输入输出
input_data, output_data = data.iloc[:, 0:2], data.iloc[:, 2]
print(input_data.shape, output_data.shape)
print(input_data.head())
print(output_data.head())
```

```python
# 处理缺失值
num_rooms_mean = input_data['NumRooms'].mean()
input_data['NumRooms'] = input_data['NumRooms'].fillna(num_rooms_mean)
print(input_data.head())
```

```python
# 处理类别型变量并确保输出为数值类型
input_encoded = pd.get_dummies(input_data, dummy_na=True)
input_encoded = input_encoded.astype('float32')  
print(input_encoded)
```

```python
# 转化为张量
x = torch.tensor(input_encoded.values)
y = torch.tensor(output_data.values.astype('float32')) 
print(x)
print(y)
```
